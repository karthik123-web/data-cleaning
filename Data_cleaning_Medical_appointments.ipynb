```python
import pandas as pd
import numpy as np
```

```python
# Load CSV file
df = pd.read_csv(r"/content/KaggleV2-May-2016.csv")

# Display first 5 rows
df.head()
```

```python
# Check shape
print("Dataset shape:", df.shape)
```

```python
# Check info (columns and types)
df.info()
```

```python
# Check for missing values in each column
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
# Check if there are duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Quick statistics for numeric and categorical columns
df.describe(include='all')
```

```python
# Drop any duplicate rows to avoid redundant data
df.drop_duplicates(inplace=True)
```

```python
# Verify duplicates are removed
print("Duplicate rows after cleaning:", df.duplicated().sum())
```

```python
# Fill missing values in 'Gender' column
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].fillna('unknown')

# Fill missing values in 'Age' column with median
if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Convert 'Gender' column to lowercase
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].str.lower()

# Convert 'No-show' column to lowercase
if 'No-show' in df.columns:
    df['No-show'] = df['No-show'].str.lower()

# Clean 'Neighbourhood' column: strip spaces and capitalize properly
if 'Neighbourhood' in df.columns:
    df['Neighbourhood'] = df['Neighbourhood'].str.strip().str.title()
```

```python
# Remove leading/trailing spaces, convert to lowercase, and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Check new column names
df.columns
```

```python
# Convert 'scheduledday' and 'appointmentday' to datetime format
if 'scheduledday' in df.columns:
    df['scheduledday'] = pd.to_datetime(df['scheduledday'], format='%Y-%m-%dT%H:%M:%SZ')

if 'appointmentday' in df.columns:
    df['appointmentday'] = pd.to_datetime(df['appointmentday'], format='%Y-%m-%dT%H:%M:%SZ')
```

```python
# Ensure 'age' column is integer type
if 'age' in df.columns:
    df['age'] = df['age'].astype(int)

# Verify types after conversion
df.info()
```

```python
# Drop any duplicate rows to avoid redundancy
df.drop_duplicates(inplace=True)

# Verify duplicates are removed
print("Duplicate rows after cleaning:", df.duplicated().sum())
```

**Saving the dataset**



```python
# Save the cleaned dataset to a new CSV file
df.to_csv("Medical_Appointment_Cleaned.csv", index=False)


```
